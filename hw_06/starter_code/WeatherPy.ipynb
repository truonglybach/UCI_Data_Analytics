### Weather Analysis of 500+ Cities for September 9, 2018
* From the data, we deduce that there is a clear negative parabolic trend, as temperature tends to increase as we come closer towards the equator.
* Intriguingly, there is a heavier concentration in cities that have 100% humidity than those that have close to 0%.
* The majority of the cities do not have winds of over 15 MPH and the most dense cluster of them have wind speeds ranging from 0-10 MPH.

```python
%matplotlib inline
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import csv
import time
from pprint import pprint

# Import API key
from config import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Make the building blocks for the query url
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
number_of_cities = len(cities)
```

```python
names_of_cities = []
lat_cities_list = []
temp_in_f_list = []
humidity_list = []
cloudiness_list = []
speed_of_wind_list = []
count = 1
timesleep_count = 0
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it's being processed (with the city number and city name).


```python
print("Beginning Data Retrieval")
print("-----------------------------")

for city in cities:
    print(f"Processing Record {count} out of {number_of_cities}: {city}")
    print(f"{url}appid={api_key}&q={city}&units=imperial")
    try:
        response = requests.get(f"{url}appid={api_key}&q={city}&units={units}")
        json_response = response.json()
        city_name = json_response["name"]
        city_lat = json_response["coord"]["lat"]
        city_temp_f = json_response["main"]["temp"]
        city_humidity = json_response["main"]["humidity"]
        city_cloudiness = json_response["clouds"]["all"]
        city_speed_of_wind_mph = json_response["wind"]["speed"]
        names_of_cities.append(city_name)
        lat_cities_list.append(city_lat)
        temp_in_f_list.append(city_temp_f)
        humidity_list.append(city_humidity)
        cloudiness_list.append(city_cloudiness)
        speed_of_wind_list.append(city_speed_of_wind_mph)
        if timesleep_count == 20:
            x = 1
            timesleep_count = 0
        else:
            x = 0
        time.sleep(x)
        count += 1
        timesleep_count += 1
    except KeyError:
        print("KeyError: Location Not Found.")
        continue

print("-----------------------------")
print(f"Data Retrieval Complete: {number_of_cities - count} cities could not be located.")
print("-----------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
dict_for_csv = {"City": names_of_cities,
"Latitude": lat_cities_list,
"Temperature in F": temp_in_f_list,
"Humidity (%)": humidity_list,
"Cloudiness (%)": cloudiness_list,
"Speed of Wind in MPH": speed_of_wind_list}

pd.DataFrame.from_dict(data=dict_for_csv).to_csv('cities_weather.csv', index=False)
```
