# Pymaceuticals Inc.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The following is an analysis of three treatments (in respect to our control, a placebo) and their effects on squamous cell carcinoma (SCC), a common form of skin cancer and the treatments are as follows:
    1. Capomulin, 
    2. Ketapril, and
    3. Infubinol.
</hr>
</hr>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The most significant of the trends from our analysis shows that Capomulin was the only drug to decrease the tumor volume over the 45-day trial, slow down the spread of metastatic sites, and prolong the largest number of lifespans of the mice. Intriguingly, in the respect of reducing tumor volume, Ketapril performed worse than the placebo, but fared better in terms of survival rate. Lastly, Infubinol performed worse than the placebo in prolonging the lifespans of the mice, yet fared better than both Ketapril and the placebo in terms of reducing tumor volume.

</hr>
</hr>
*<i>Data visualizations are listed below</i>.

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_og_df = pd.read_csv(mouse_drug_data_to_load)
clinical_og_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
mouse_og_df = mouse_og_df.sort_values(by="Mouse ID")
clinical_og_df = clinical_og_df.sort_values(by="Mouse ID")
combined_df = pd.merge(mouse_og_df, clinical_og_df, on="Mouse ID", how="outer")

# Display the data table for preview
combined_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_drug_timepoint = combined_df.groupby(["Drug", "Timepoint"])
tumor_vol_mean = mean_drug_timepoint["Tumor Volume (mm3)"].mean()
# Convert to DataFrame
df_tumor_vol = pd.DataFrame(tumor_vol_mean)
#Preview DataFrame
df_tumor_vol.head()
```

```python
# Store the Standard Error of Tumor Volumes
SE_tv = combined_df.groupby(["Drug", "Timepoint"]).sem()
SE_tv = SE_tv.rename(columns={"Tumor Volume (mm3)": "SE of Tumor Vol"})
del SE_tv["Mouse ID"]
# Preview DataFrame
SE_tv.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatted_tumor_response_df = df_tumor_vol.reset_index().pivot(index="Timepoint", columns="Drug")
# Preview that Reformatting worked
reformatted_tumor_response_df
```

```python
# Generate the Plot (with Error Bars)
plt.figure(figsize=(10,6))
x_axis = np.arange(0, 10 ,1)
x_tick_labels = combined_df["Timepoint"].sort_values().unique()
# Get y-values for Capomulin
capomulin_values = combined_df.loc[combined_df["Drug"] == "Capomulin"]
y_axis_capomulin = capomulin_values.groupby("Timepoint").mean().rename(columns={"Tumor Volume (mm3)": "Capomulin"})["Capomulin"]
# Reset SE dataframe
reset_SE = SE_tv.reset_index(level=0)
# Get SE values for Capomulin
capo_SE_df = reset_SE.loc[reset_SE["Drug"] == "Capomulin", :]
capo_SE = capo_SE_df['SE of Tumor Vol']
# Plot the points
capomulin_chart = plt.errorbar(x_axis, y_axis_capomulin, capo_SE, fmt="o", color="#138E3E", elinewidth=2, marker='*')
# Find line of best fit
(slope_cap, intercept_cap, r_value_cap, p_value_cap, std_err_cap) = linregress(x_axis, y_axis_capomulin)
capo_fit = slope_cap * x_axis + intercept_cap


# Get y-values for Placebo
placebo_values = combined_df.loc[combined_df["Drug"] == "Placebo"]
y_axis_placebo = placebo_values.groupby("Timepoint").mean().rename(columns={"Tumor Volume (mm3)": "Placebo"})["Placebo"]
# Get SE values for Placebo
placebo_SE_df = reset_SE.loc[reset_SE["Drug"] == "Placebo", :]
placebo_SE = placebo_SE_df["SE of Tumor Vol"]
# Plot the points
placebo_chart = plt.errorbar(x_axis, y_axis_placebo, placebo_SE, fmt="o", color="#FF0000", elinewidth=2, marker='s')
# Find line of best fit
(slope_pb, intercept_pb, r_value_pb, p_value_pb, std_err_pb) = linregress(x_axis, y_axis_placebo)
pb_fit = slope_pb * x_axis + intercept_pb

# Get y-values for Ketapril
ketapril_values = combined_df.loc[combined_df["Drug"] == "Ketapril"]
y_axis_ketapril = ketapril_values.groupby("Timepoint").mean().rename(columns={"Tumor Volume (mm3)": "Ketapril"})["Ketapril"]
# Get SE values for Ketapril
ketapril_SE_df = reset_SE.loc[reset_SE["Drug"] == "Ketapril", :]
ketapril_SE = ketapril_SE_df["SE of Tumor Vol"]
# Plot the points
ketapril_chart = plt.errorbar(x_axis, y_axis_ketapril, ketapril_SE, fmt="o", color="#0B00F8", elinewidth=2, marker='v')
# Find line of best fit
(slope_kt, intercept_kt, r_value_kt, p_value_kt, std_err_kt) = linregress(x_axis, y_axis_ketapril)
kt_fit = slope_kt * x_axis + intercept_kt

# Get y-values for Infubinol
infubinol_values = combined_df.loc[combined_df["Drug"] == "Infubinol"]
y_axis_infubinol = infubinol_values.groupby("Timepoint").mean().rename(columns={"Tumor Volume (mm3)": "Infubinol"})["Infubinol"]
# Get SE values for Infubinol
infubinol_SE_df = reset_SE.loc[reset_SE["Drug"] == "Infubinol", :]
infubinol_SE = infubinol_SE_df["SE of Tumor Vol"]
# Plot the points
infubinol_chart = plt.errorbar(x_axis, y_axis_infubinol, infubinol_SE, fmt="o", color="#A409E7", elinewidth=2, marker='+')
# Find line of best fit
(slope_rm, intercept_rm, r_value_rm, p_value_rm, std_err_rm) = linregress(x_axis, y_axis_infubinol)
if_fit = slope_rm * x_axis + intercept_rm

# Draw the lines of best fit and put up the legends
plt.legend(loc='best')
plotted_drugs = plt.plot(x_axis, capo_fit, '#138E3E', linestyle='dashed'), plt.plot(x_axis, pb_fit, '#FF0000', linestyle='dashed'), plt.plot(x_axis, kt_fit, '#0B00F8', linestyle='dashed'), plt.plot(x_axis, if_fit, '#A409E7', linestyle='dashed')

plt.xlabel("Number of Days")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")
plt.xticks(x_axis, x_tick_labels)
plt.grid()
plt.xlim(-.5, len(x_axis)-.5)
plt.tight_layout()

# Save the Figure
plt.savefig("tumor_response.png", dpi=300)

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_s = combined_df.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()

# Convert to DataFrame
mean_met_df = pd.DataFrame(mean_met_s)

# Preview DataFrame
mean_met_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_met_s = combined_df.groupby(["Drug", "Timepoint"])["Metastatic Sites"].sem()
# Convert to DataFrame
sem_met_df = pd.DataFrame(sem_met_s)
# Preview DataFrame
sem_met_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatted_met_df = mean_met_df.reset_index().pivot(index="Timepoint", columns="Drug")
# Preview that Reformatting worked
reformatted_met_df
```

```python
# Generate the Plot (with Error Bars)
plt.figure(figsize=(10,6))

# Find mean Met values for Capomulin and store as drug name
y_met_mean_capo = capomulin_values.groupby("Timepoint").mean().rename(columns={"Metastatic Sites": "Capomulin"})["Capomulin"]
# Get SE values for Capomulin Met sites
capo_met_SE = capo_SE_df['Metastatic Sites']
# Plot the points
capo_met_chart = plt.errorbar(x_axis, y_met_mean_capo, capo_met_SE, fmt='o', color="#138E3E", elinewidth=2, marker='*')
# Find the line of best fit
(slope_cap_met, intercept_cap_met, r_value_cap_met, p_value_cap_met, std_err_cap_met) = linregress(x_axis, y_met_mean_capo)
y_cap_met_lfit = slope_cap_met * x_axis + intercept_cap_met

# Find mean Met values for Placebo and store as drug name
y_met_mean_pb = placebo_values.groupby("Timepoint").mean().rename(columns={"Metastatic Sites": "Placebo"})["Placebo"]
# Get SE values for Placebo Met sites
pb_met_SE = placebo_SE_df['Metastatic Sites']
# Plot the points
pb_met_chart = plt.errorbar(x_axis, y_met_mean_pb, pb_met_SE, fmt='o', color='#FF0000', elinewidth=2, marker='s')
# Find the line of best fit
(slope_pb_met, intercept_pb_met, r_value_pb_met, p_value_pb_met, std_err_pb_met) = linregress(x_axis, y_met_mean_pb)
y_pb_met_lfit = slope_pb_met * x_axis + intercept_pb_met

# Find mean Met values for Ketapril and store as drug name
y_met_mean_kt = ketapril_values.groupby("Timepoint").mean().rename(columns={"Metastatic Sites": "Ketapril"})["Ketapril"]
# Get SE values for Ketapril Met sites
kt_met_SE = ketapril_SE_df['Metastatic Sites']
# Plot the points
kt_met_chart = plt.errorbar(x_axis, y_met_mean_kt, kt_met_SE, fmt='o', color="#0B00F8", elinewidth=2, marker='v')
# Find the line of best fit
(slope_kt_met, intercept_kt_met, r_value_kt_met, p_value_kt_met, std_err_kt_met) = linregress(x_axis, y_met_mean_kt)
y_kt_met_lfit = slope_kt_met * x_axis + intercept_kt_met

# Find mean Met values for Infubinol and store as drug name
y_met_mean_if = infubinol_values.groupby("Timepoint").mean().rename(columns={"Metastatic Sites": "Infubinol"})["Infubinol"]
# Get SE values for Infubinol Met sites
if_met_SE = infubinol_SE_df['Metastatic Sites']
# Plot the points
if_met_chart = plt.errorbar(x_axis, y_met_mean_if, if_met_SE, fmt='o', color="#A409E7", elinewidth=2, marker='+')
# Find the line of best fit
(slope_if_met, intercept_if_met, r_value_if_met, p_value_if_met, std_err_if_met) = linregress(x_axis, y_met_mean_if)
y_if_met_lfit = slope_if_met * x_axis + intercept_if_met

# Draw the lines of best fit and put up the legends
plt.legend(loc='best')
plotted_drugs = plt.plot(x_axis, y_cap_met_lfit, '#138E3E', linestyle='dashed'), plt.plot(x_axis, y_pb_met_lfit, '#FF0000', linestyle='dashed'), plt.plot(x_axis, y_kt_met_lfit, '#0B00F8', linestyle='dashed'), plt.plot(x_axis, y_if_met_lfit, '#A409E7', linestyle='dashed')

plt.xlabel("Number of Days")
plt.ylabel("Metastatic Sites")
plt.title("Metastatic Response to Treatment")
plt.xticks(x_axis, x_tick_labels)
plt.grid()

plt.tight_layout()

# Save the Figure
plt.savefig("metastatic_response.png", dpi=300)

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_rate_count = combined_df.groupby(["Drug", "Timepoint"]).count().rename(columns={"Mouse ID": "Mouse Count"})["Mouse Count"]
# Convert to DataFrame
survival_rate_count_df = pd.DataFrame(survival_rate_count)
# Preview DataFrame
survival_rate_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatted_SR_df = survival_rate_count_df.reset_index().pivot(index="Timepoint", columns="Drug")
# Preview the Data Frame
reformatted_SR_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.figure(figsize=(10,6))

# Find SR points of Capomulin mice group
y_SR_count_capo = capomulin_values.groupby("Timepoint").count().rename(columns={"Mouse ID": "Capomulin"})["Capomulin"]
# Plot the points
capo_SR_chart = plt.errorbar(x_axis, y_SR_count_capo, fmt='o', color="#138E3E", elinewidth=2, marker='*')
# Find the line of best fit
(slope_cap_SR, intercept_cap_SR, r_value_cap_SR, p_value_cap_SR, std_err_cap_SR) = linregress(x_axis, y_SR_count_capo)
y_cap_SR_lfit = slope_cap_SR * x_axis + intercept_cap_SR

# Find SR points of Placebo mice group
y_SR_count_pb = placebo_values.groupby("Timepoint").count().rename(columns={"Mouse ID": "Placebo"})["Placebo"]
# Plot the points
pb_SR_chart = plt.errorbar(x_axis, y_SR_count_pb, fmt='o', color='#FF0000', elinewidth=2, marker='s')
# Find the line of best fit
(slope_pb_SR, intercept_pb_SR, r_value_pb_SR, p_value_pb_SR, std_err_pb_SR) = linregress(x_axis, y_SR_count_pb)
y_pb_SR_lfit = slope_pb_SR * x_axis + intercept_pb_SR

# Find SR points of Ketapril mice group
y_SR_count_kt = ketapril_values.groupby("Timepoint").count().rename(columns={"Mouse ID": "Ketapril"})["Ketapril"]
# Plot the points
ket_SR_chart = plt.errorbar(x_axis, y_SR_count_kt, fmt='o', color="#0B00F8", elinewidth=2, marker='v')
# Find the line of best fit
(slope_kt_SR, intercept_kt_SR, r_value_kt_SR, p_value_kt_SR, std_err_kt_SR) = linregress(x_axis, y_SR_count_kt)
y_kt_SR_lfit = slope_kt_SR * x_axis + intercept_kt_SR

# Find SR points of Infubinol mice group
y_SR_count_if = infubinol_values.groupby("Timepoint").count().rename(columns={"Mouse ID": "Infubinol"})["Infubinol"]
# Plot the points
if_SR_chart = plt.errorbar(x_axis, y_SR_count_if, fmt='o', color="#A409E7", elinewidth=2, marker='+')
# Find the line of best fit
(slope_if_SR, intercept_if_SR, r_value_if_SR, p_value_if_SR, std_err_if_SR) = linregress(x_axis, y_SR_count_if)
y_if_SR_lfit = slope_if_SR * x_axis + intercept_if_SR

# Draw the lines of best fit and put up the legends
plt.legend(loc='best')
plotted_drugs = plt.plot(x_axis, y_cap_SR_lfit, '#138E3E', linestyle='dashed'), plt.plot(x_axis, y_pb_SR_lfit, '#FF0000', linestyle='dashed'), plt.plot(x_axis, y_kt_SR_lfit, '#0B00F8', linestyle='dashed'), plt.plot(x_axis, y_if_SR_lfit, '#A409E7', linestyle='dashed')

plt.xlabel("Number of Days")
plt.ylabel("Number of Mice")
plt.title("Survival Rates")
plt.xticks(x_axis, x_tick_labels)
plt.grid()

plt.tight_layout()

# Save the Figure
plt.savefig('survival_rates.png', dpi=300)

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Reset the tumor volume data frame
reset_per_df = df_tumor_vol.reset_index()

# Store all Relevant Percent Changes into a Tuple
per_array = []
# Store drug labels for bar chart
drug_labels = []

# Locate all of the data for Capomulin and store the value
cap_per_df = reset_per_df.loc[reset_per_df["Drug"] == "Capomulin"]
# Get the drug names to use as labels
cap_name = cap_per_df.iloc[0, 0]
drug_labels.append(cap_name)
# Find the initial and final value for tumor volume
min_cap_per = cap_per_df.loc[cap_per_df["Timepoint"] == 0, :].iloc[0, 2]
max_cap_per = cap_per_df.loc[cap_per_df["Timepoint"] == 45, :].iloc[0, 2]
# Calculate the percent changes for each drug
rounded_per_cap = round((((max_cap_per - min_cap_per) / min_cap_per) * 100), 2)
per_array.append(rounded_per_cap)

# Locate all of the data for the Placebo and store the value
pb_per_df = reset_per_df.loc[reset_per_df["Drug"] == "Placebo"]
# Get the drug names to use as labels
pb_name = pb_per_df.iloc[0, 0]
drug_labels.append(pb_name)
# Find the initial and final value for tumor volume
min_pb_per = pb_per_df.loc[pb_per_df["Timepoint"] == 0, :].iloc[0, 2]
max_pb_per = pb_per_df.loc[pb_per_df["Timepoint"] == 45, :].iloc[0, 2]
# Calculate the percent changes for each drug
rounded_per_pb = round((((max_pb_per - min_pb_per) / min_pb_per) * 100), 2)
per_array.append(rounded_per_pb)

# Locate all of the data for Ketapril and store the value
kt_per_df = reset_per_df.loc[reset_per_df["Drug"] == "Ketapril"]
# Get the drug names to use as labels
kt_name = kt_per_df.iloc[0, 0]
drug_labels.append(kt_name)
# Find the initial and final value for tumor volume
min_kt_per = kt_per_df.loc[kt_per_df["Timepoint"] == 0, :].iloc[0, 2]
max_kt_per = kt_per_df.loc[kt_per_df["Timepoint"] == 45, :].iloc[0, 2]
# Calculate the percent changes for each drug
rounded_per_kt = round((((max_kt_per - min_kt_per) / min_kt_per) * 100), 2)
per_array.append(rounded_per_kt)

# Locate all of the data for Infubinol and store the value
if_per_df = reset_per_df.loc[reset_per_df["Drug"] == "Infubinol"]
# Get the drug names to use as labels
if_name = if_per_df.iloc[0, 0]
drug_labels.append(if_name)
# Find the initial and final value for tumor volume
min_if_per = if_per_df.loc[if_per_df["Timepoint"] == 0, :].iloc[0, 2]
max_if_per = if_per_df.loc[if_per_df["Timepoint"] == 45, :].iloc[0, 2]
# Calculate the percent changes for each drug
rounded_per_if = round((((max_if_per - min_if_per) / min_if_per) * 100), 2)
per_array.append(rounded_per_if)

# Display the data
plt.figure(figsize=(10, 6))
bar_chart = plt.bar((range(len(per_array))), per_array, align='center', tick_label=drug_labels, color=('#06940D', '#BC0808', '#BC0808', '#BC0808'))
plt.xlabel("Drug")
plt.ylabel("Percentage Change over 45 Days")
plt.title("Percentage Change of Tumor Volume")

# Orient widths. Add labels, tick marks, etc. 
plt.hlines(0, -0.5, 3.5)
plt.ylim((min(per_array)-5), (max(per_array)+5))
per_change_chart = plt.xlim((min(range(len(per_array))))-.5, (max(range(len(per_array))))+.5)
plt.tight_layout()
plt.grid()

# Save figure
plt.savefig("per_change(tumor_vol)_barchart.png", dpi=300)

# Show the figure
plt.show()
```
